# Have to select what K will use for the maximum accuaracy 

# Lets write a function for it

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import numpy as np
error_rate= []
for i in range(1,40):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,encoded_y)
    pred= knn.predict(x_test)
    error_rate.append(np.mean(pred != encoded_ytest))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o', markerfacecolor='red', markersize=10)
plt.xlabel('k value')
plt.ylabel('Error rate')
plt.title('To check the correct value of k')
plt.show()
           
```

<span style="font-size:24px;">from the graph should select K=12 as it gives the best error rate</span>
